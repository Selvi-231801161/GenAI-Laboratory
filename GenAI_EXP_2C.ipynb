```python
!pip install gensim
import gensim.downloader as api
import numpy as np
from numpy.linalg import norm

model = api.load("word2vec-google-news-300")

def cosine(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))

def association(w, A, B):
    return np.mean([cosine(model[w], model[a]) for a in A]) - \
           np.mean([cosine(model[w], model[b]) for b in B])

def test_statistic(X, Y, A, B):
    return sum(association(x, A, B) for x in X) - \
           sum(association(y, A, B) for y in Y)

def effect_size(X, Y, A, B):
    assoc_X = [association(x, A, B) for x in X]
    assoc_Y = [association(y, A, B) for y in Y]

    mean_diff = np.mean(assoc_X) - np.mean(assoc_Y)
    std_dev = np.std(assoc_X + assoc_Y)

    return mean_diff / std_dev

career = ["executive", "management", "professional", "corporation", "salary"]
family = ["home", "parents", "children"]

male = ["man", "male", "boy", "brother", "he"]
female = ["woman", "female", "girl", "sister"]

science = ["science", "technology", "physics"]
arts = ["poetry", "art", "dance", "literature", "novel"]

stat1 = test_statistic(career, family, male, female)
effect1 = effect_size(career, family, male, female)

print("Career vs Family bias:")
print("Test Statistic:", stat1)
print("Effect Size:", effect1)

print("\nGender Bias (Science vs Arts):")
stat2 = test_statistic(science, arts, male, female)
effect2 = effect_size(science, arts, male, female)

print("Test Statistic:", stat2)
print("Effect Size:", effect2)

if effect1 > 0:
    print("\nObservation: Career words are more associated with male.")
else:
    print("\nObservation: No strong gender association detected (career-family).")

if effect2 > 0:
    print("Observation: Science words are more associated with male.")
else:
    print("Observation: No strong gender association detected (science-arts).")
```
